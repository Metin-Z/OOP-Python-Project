```python
import tkinter as tk
from tkinter import messagebox

class Bilet:
    def __init__(self, bilet_id):
        self.bilet_id = bilet_id

class Katilimci:
    def __init__(self, isim, soyisim, bilet_id):
        self.isim = isim
        self.soyisim = soyisim
        self.bilet_id = bilet_id

class Etkinlik:
    def __init__(self, etkinlik_adi, tarih):
        self.etkinlik_adi = etkinlik_adi
        self.tarih = tarih
        self.katilimcilar = []

    def katilimci_ekle(self, katilimci):
        self.katilimcilar.append(katilimci)

class EtkinlikYonetimSistemi:
    def __init__(self):
        self.etkinlikler = []

    def etkinlik_ekle(self, etkinlik):
        self.etkinlikler.append(etkinlik)

    def etkinlik_bul(self, etkinlik_adi):
        for etkinlik in self.etkinlikler:
            if etkinlik.etkinlik_adi == etkinlik_adi:
                return etkinlik
        return None

class EtkinlikArayuzu:
    def __init__(self, etkinlik_yonetim_sistemi):
        self.etkinlik_yonetim_sistemi = etkinlik_yonetim_sistemi
        self.ana_pencere = tk.Tk()
        self.ana_pencere.title("Etkinlik Yönetim Sistemi")

        self.etkinlik_secim_cerceve = tk.LabelFrame(self.ana_pencere, text="Etkinlik Ekle")
        self.etkinlik_secim_cerceve.pack(fill="both", expand="yes", padx=20, pady=10)

        self.etkinlik_adi_label = tk.Label(self.etkinlik_secim_cerceve, text="Etkinlik Adı:")
        self.etkinlik_adi_label.grid(row=0, column=0)
        self.etkinlik_adi_entry = tk.Entry(self.etkinlik_secim_cerceve)
        self.etkinlik_adi_entry.grid(row=0, column=1)

        self.etkinlik_tarih_label = tk.Label(self.etkinlik_secim_cerceve, text="Tarih:")
        self.etkinlik_tarih_label.grid(row=1, column=0)
        self.etkinlik_tarih_entry = tk.Entry(self.etkinlik_secim_cerceve)
        self.etkinlik_tarih_entry.grid(row=1, column=1)

        self.etkinlik_isim_label = tk.Label(self.etkinlik_secim_cerceve, text="İsim:")
        self.etkinlik_isim_label.grid(row=2, column=0)
        self.etkinlik_isim_entry = tk.Entry(self.etkinlik_secim_cerceve)
        self.etkinlik_isim_entry.grid(row=2, column=1)

        self.etkinlik_soyisim_label = tk.Label(self.etkinlik_secim_cerceve, text="Soyisim:")
        self.etkinlik_soyisim_label.grid(row=3, column=0)
        self.etkinlik_soyisim_entry = tk.Entry(self.etkinlik_secim_cerceve)
        self.etkinlik_soyisim_entry.grid(row=3, column=1)

        self.etkinlik_bilet_label = tk.Label(self.etkinlik_secim_cerceve, text="Bilet ID:")
        self.etkinlik_bilet_label.grid(row=4, column=0)
        self.etkinlik_bilet_entry = tk.Entry(self.etkinlik_secim_cerceve)
        self.etkinlik_bilet_entry.grid(row=4, column=1)

        self.ekle_button = tk.Button(self.etkinlik_secim_cerceve, text="Etkinlik Ekle", command=self.etkinlik_ekle)
        self.ekle_button.grid(row=5, columnspan=2)

    def etkinlik_ekle(self):
        etkinlik_adi = self.etkinlik_adi_entry.get()
        tarih = self.etkinlik_tarih_entry.get()
        isim = self.etkinlik_isim_entry.get()
        soyisim = self.etkinlik_soyisim_entry.get()
        bilet_id = self.etkinlik_bilet_entry.get()
        if etkinlik_adi and tarih and isim and soyisim and bilet_id:
            etkinlik = self.etkinlik_yonetim_sistemi.etkinlik_bul(etkinlik_adi)
            if etkinlik:
                etkinlik.katilimci_ekle(Katilimci(isim, soyisim, bilet_id))
                messagebox.showinfo("Başarılı", f"'{etkinlik_adi}' etkinliğine katılım sağlandı.")
            else:
                messagebox.showerror("Hata", f"'{etkinlik_adi}' etkinliği bulunamadı.")
        else:
            messagebox.showerror("Hata", "Lütfen tüm alanları doldurun.")

if __name__ == "__main__":
    etkinlik_yonetim_sistemi = EtkinlikYonetimSistemi()
    etkinlik_yonetim_sistemi.etkinlik_ekle(Etkinlik("Deep Purple Konseri", "2024-05-15"))
    etkinlik_yonetim_sistemi.etkinlik_ekle(Etkinlik("Andre Rieu Konseri", "2024-06-20"))
    etkinlik_yonetim_sistemi.etkinlik_ekle(Etkinlik("Duman Konseri", "2024-07-10"))
    arayuz = EtkinlikArayuzu(etkinlik_yonetim_sistemi)
    arayuz.ana_pencere.mainloop()
```
