```python
import tkinter as tk
from tkinter import messagebox

class Urun:
    def __init__(self, urun_adi, stok_miktari):
        self.urun_adi = urun_adi
        self.stok_miktari = stok_miktari

class Stok:
    def __init__(self):
        self.urunler = []

    def urun_ekle(self, urun):
        self.urunler.append(urun)

    def stok_guncelle(self, urun_adi, yeni_stok_miktari):
        for urun in self.urunler:
            if urun.urun_adi.lower() == urun_adi.lower():
                urun.stok_miktari = yeni_stok_miktari
                return True
        return False

    def stok_durumu_goster(self):
        stok_durumu = ""
        for urun in self.urunler:
            stok_durumu += f"{urun.urun_adi}: {urun.stok_miktari}\n"
        messagebox.showinfo("Stok Durumu", stok_durumu)

class Siparis:
    def __init__(self, siparis_no):
        self.siparis_no = siparis_no
        self.siparis_detaylari = []

    def siparis_detay_ekle(self, urun_adi, miktar):
        self.siparis_detaylari.append((urun_adi, miktar))

    def siparis_detay_sil(self, index):
        del self.siparis_detaylari[index]

class SiparisArayuzu:
    def __init__(self, stok_sistemi):
        self.stok_sistemi = stok_sistemi
        self.mevcut_siparisler = []

        self.ana_pencere = tk.Tk()
        self.ana_pencere.title("Üst Düzey Şirket Sipariş Yönetim Sistemi")
        self.ana_pencere.geometry("800x600")

        self.ana_pencere.configure(bg="#6495ED")

        self.siparis_secim_cerceve = tk.LabelFrame(self.ana_pencere, text="Sipariş Oluştur", bg="#87CEEB", padx=20, pady=20)
        self.siparis_secim_cerceve.place(relx=0.05, rely=0.05, relwidth=0.4, relheight=0.3)

        self.urun_adi_label = tk.Label(self.siparis_secim_cerceve, text="Ürün Adı:", bg="#87CEEB", padx=5, pady=5)
        self.urun_adi_label.grid(row=0, column=0, sticky="w")

        self.urun_adi_entry = tk.Entry(self.siparis_secim_cerceve)
        self.urun_adi_entry.grid(row=0, column=1, padx=5, pady=5)

        self.miktar_label = tk.Label(self.siparis_secim_cerceve, text="Miktar:", bg="#87CEEB", padx=5, pady=5)
        self.miktar_label.grid(row=1, column=0, sticky="w")

        self.miktar_entry = tk.Entry(self.siparis_secim_cerceve)
        self.miktar_entry.grid(row=1, column=1, padx=5, pady=5)

        self.siparis_no_label = tk.Label(self.siparis_secim_cerceve, text="Sipariş No:", bg="#87CEEB", padx=5, pady=5)
        self.siparis_no_label.grid(row=2, column=0, sticky="w")

        self.siparis_no_entry = tk.Entry(self.siparis_secim_cerceve)
        self.siparis_no_entry.grid(row=2, column=1, padx=5, pady=5)

        self.ekle_button = tk.Button(self.siparis_secim_cerceve, text="Sipariş Oluştur", command=self.siparis_olustur, bg="#32CD32", fg="white", padx=10, pady=5)
        self.ekle_button.grid(row=3, columnspan=2, pady=10)

        self.mevcut_siparisler_button = tk.Button(self.ana_pencere, text="Mevcut Siparişleri Göster", command=self.mevcut_siparisleri_goster, bg="#87CEEB", padx=10, pady=5)
        self.mevcut_siparisler_button.place(relx=0.05, rely=0.45)

        self.stok_durumu_cerceve = tk.LabelFrame(self.ana_pencere, text="Stok Durumu", bg="#87CEEB", padx=20, pady=20)
        self.stok_durumu_cerceve.place(relx=0.5, rely=0.05, relwidth=0.4, relheight=0.9)

        self.stok_durumu_guncelle()

        self.stok_guncelle_button = tk.Button(self.ana_pencere, text="Stok Güncelle", command=self.stok_guncelle_arayuzunu_ac, bg="#32CD32", fg="white", padx=10, pady=5)
        self.stok_guncelle_button.place(relx=0.75, rely=0.90, anchor="center")

    def stok_guncelle_arayuzunu_ac(self):
        stok_guncelleme_pencere = tk.Toplevel(self.ana_pencere)
        stok_guncelleme_pencere.title("Stok Güncelleme")
        stok_guncelleme_pencere.geometry("300x200")

        urun_adi_label = tk.Label(stok_guncelleme_pencere, text="Ürün Adı:", padx=5, pady=5)
        urun_adi_label.grid(row=0, column=0)

        urun_adi_entry = tk.Entry(stok_guncelleme_pencere)
        urun_adi_entry.grid(row=0, column=1, padx=5, pady=5)

        miktar_label = tk.Label(stok_guncelleme_pencere, text="Yeni Stok Miktarı:", padx=5, pady=5)
        miktar_label.grid(row=1, column=0)

        miktar_entry = tk.Entry(stok_guncelleme_pencere)
        miktar_entry.grid(row=1, column=1, padx=5, pady=5)

        guncelle_button = tk.Button(stok_guncelleme_pencere, text="Güncelle", command=lambda: self.stok_guncelle_onay(urun_adi_entry.get(), miktar_entry.get(), stok_guncelleme_pencere), bg="#32CD32", fg="white", padx=10, pady=5)
        guncelle_button.grid(row=2, columnspan=2, pady=10)

    def stok_guncelle_onay(self, urun_adi, yeni_stok_miktari, pencere):
        if self.stok_sistemi.stok_guncelle(urun_adi, int(yeni_stok_miktari)):
            messagebox.showinfo("Başarılı", f"'{urun_adi}' stok miktarı başarıyla güncellendi.")
            self.stok_durumu_guncelle()
        else:
            messagebox.showerror("Hata", f"'{urun_adi}' ürünü bulunamadı.")
        pencere.destroy()

    def stok_durumu_guncelle(self):
        for widget in self.stok_durumu_cerceve.winfo_children():
            widget.destroy()

        self.stok_durumu_label = tk.Label(self.stok_durumu_cerceve, text="Stok Durumu", bg="#87CEEB", padx=5, pady=5, font=("Helvetica", 16))
        self.stok_durumu_label.pack()

        for i, urun in enumerate(self.stok_sistemi.urunler):
            urun_label = tk.Label(self.stok_durumu_cerceve, text=f"{urun.urun_adi}: {urun.stok_miktari}", bg="#87CEEB", padx=5, pady=5)
            urun_label.pack(anchor="w")

    def siparis_olustur(self):
        urun_adi = self.urun_adi_entry.get()
        miktar = int(self.miktar_entry.get())
        siparis_no = self.siparis_no_entry.get()

        urun_var = False
        for urun in self.stok_sistemi.urunler:
            if urun.urun_adi.lower() == urun_adi.lower():
                urun_var = True
                if urun.stok_miktari >= miktar:
                    siparis = Siparis(siparis_no)
                    siparis.siparis_detay_ekle(urun_adi, miktar)
                    self.mevcut_siparisler.append(siparis)
                    messagebox.showinfo("Başarılı", f"'{siparis_no}' numaralı sipariş başarıyla oluşturuldu.")
                    urun.stok_miktari -= miktar
                    self.stok_durumu_guncelle()
                else:
                    messagebox.showerror("Hata", f"Stokta yeterli '{urun_adi}' bulunmamaktadır. Stoktaki miktar: {urun.stok_miktari}")
                break

        if not urun_var:
            messagebox.showerror("Hata", f"'{urun_adi}' ürünü bulunamadı.")

    def mevcut_siparisleri_goster(self):
        self.mevcut_siparisler_pencere = tk.Toplevel(self.ana_pencere)
        self.mevcut_siparisler_pencere.title("Mevcut Siparişler")
        self.mevcut_siparisler_pencere.geometry("400x300")

        mevcut_siparisler_listbox = tk.Listbox(self.mevcut_siparisler_pencere, width=50, height=15)
        mevcut_siparisler_listbox.pack(padx=10, pady=10)

        for siparis in self.mevcut_siparisler:
            for detay in siparis.siparis_detaylari:
                urun_adi, miktar = detay
                mevcut_siparisler_listbox.insert(tk.END, f"Sipariş No: {siparis.siparis_no}, Ürün: {urun_adi}, Miktar: {miktar}")
                sil_button = tk.Button(self.mevcut_siparisler_pencere, text="Siparişi Sil", command=lambda siparis=siparis: self.siparisi_sil(siparis, mevcut_siparisler_listbox), bg="#DC143C", fg="white", padx=10, pady=5)
                sil_button.pack(pady=5)

    def siparisi_sil(self, siparis, listbox):
        siparis_index = self.mevcut_siparisler.index(siparis)
        del self.mevcut_siparisler[siparis_index]
        listbox.delete(siparis_index)
        messagebox.showinfo("Başarılı", "Sipariş başarıyla silindi.")
        self.stok_durumu_guncelle()

if __name__ == "__main__":
    stok_sistemi = Stok()
    stok_sistemi.urun_ekle(Urun("Yumurta", 200))
    stok_sistemi.urun_ekle(Urun("Un", 150))
    stok_sistemi.urun_ekle(Urun("Şeker", 250))
    stok_sistemi.urun_ekle(Urun("Tuz", 300))
    stok_sistemi.urun_ekle(Urun("Süt", 100))
    stok_sistemi.urun_ekle(Urun("Makarna", 180))
    stok_sistemi.urun_ekle(Urun("Pirinç", 220))
    stok_sistemi.urun_ekle(Urun("Domates", 160))
    stok_sistemi.urun_ekle(Urun("Salatalık", 200))
    stok_sistemi.urun_ekle(Urun("Patates", 180))

    arayuz = SiparisArayuzu(stok_sistemi)
    arayuz.ana_pencere.mainloop()
```
