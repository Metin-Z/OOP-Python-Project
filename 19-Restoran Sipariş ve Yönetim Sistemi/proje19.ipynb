```python
import tkinter as tk
from tkinter import messagebox

class Urun:
    def __init__(self, urun_adi, stok_miktari, fiyat):
        self.urun_adi = urun_adi
        self.stok_miktari = stok_miktari
        self.fiyat = fiyat

class Stok:
    def __init__(self):
        self.urunler = []

    def urun_ekle(self, urun):
        self.urunler.append(urun)

    def stok_guncelle(self, urun_adi, yeni_stok_miktari, yeni_fiyat):
        for urun in self.urunler:
            if urun.urun_adi.lower() == urun_adi.lower():
                urun.stok_miktari = yeni_stok_miktari
                urun.fiyat = yeni_fiyat
                return True
        return False

    def stok_durumu_goster(self):
        stok_durumu = ""
        for urun in self.urunler:
            stok_durumu += f"{urun.urun_adi}: Stok: {urun.stok_miktari}, Fiyat: {urun.fiyat} TL\n"
        messagebox.showinfo("Stok Durumu", stok_durumu)

class Siparis:
    def __init__(self, siparis_no, musteri_ad, adres):
        self.siparis_no = siparis_no
        self.musteri_ad = musteri_ad
        self.adres = adres
        self.siparis_detaylari = []

    def siparis_detay_ekle(self, urun_adi):
        self.siparis_detaylari.append(urun_adi)

    def siparis_detay_sil(self, index):
        del self.siparis_detaylari[index]

class SiparisArayuzu:
    def __init__(self, stok_sistemi):
        self.stok_sistemi = stok_sistemi
        self.mevcut_siparisler = []

        self.ana_pencere = tk.Tk()
        self.ana_pencere.title("Üst Düzey Şirket Sipariş Yönetim Sistemi")
        self.ana_pencere.geometry("800x600")

        self.ana_pencere.configure(bg="#FF69B4")

        self.siparis_secim_cerceve = tk.LabelFrame(self.ana_pencere, text="Sipariş Oluştur", bg="#FFFF00", padx=20, pady=20)
        self.siparis_secim_cerceve.place(relx=0.05, rely=0.05, relwidth=0.4, relheight=0.4)

        self.urun_adi_label = tk.Label(self.siparis_secim_cerceve, text="Yemek Adı:", bg="#FFFF00", padx=5, pady=5)
        self.urun_adi_label.grid(row=0, column=0, sticky="w")

        self.urun_adi_entry = tk.Entry(self.siparis_secim_cerceve)
        self.urun_adi_entry.grid(row=0, column=1, padx=5, pady=5)

        self.adres_label = tk.Label(self.siparis_secim_cerceve, text="Adres:", bg="#FFFF00", padx=5, pady=5)
        self.adres_label.grid(row=1, column=0, sticky="w")

        self.adres_entry = tk.Entry(self.siparis_secim_cerceve)
        self.adres_entry.grid(row=1, column=1, padx=5, pady=5)

        self.musteri_ad_label = tk.Label(self.siparis_secim_cerceve, text="Müşteri Adı:", bg="#FFFF00", padx=5, pady=5)
        self.musteri_ad_label.grid(row=2, column=0, sticky="w")

        self.musteri_ad_entry = tk.Entry(self.siparis_secim_cerceve)
        self.musteri_ad_entry.grid(row=2, column=1, padx=5, pady=5)

        self.siparis_no_label = tk.Label(self.siparis_secim_cerceve, text="Sipariş No:", bg="#FFFF00", padx=5, pady=5)
        self.siparis_no_label.grid(row=3, column=0, sticky="w")

        self.siparis_no_entry = tk.Entry(self.siparis_secim_cerceve)
        self.siparis_no_entry.grid(row=3, column=1, padx=5, pady=5)

        self.ekle_button = tk.Button(self.siparis_secim_cerceve, text="Sipariş Oluştur", command=self.siparis_olustur, bg="#4169E1", fg="white", padx=10, pady=5)
        self.ekle_button.grid(row=4, columnspan=2, pady=10)

        self.mevcut_siparisler_button = tk.Button(self.ana_pencere, text="Mevcut Siparişleri Göster", command=self.mevcut_siparisleri_goster, bg="#FFFF00", padx=10, pady=5)
        self.mevcut_siparisler_button.place(relx=0.05, rely=0.55)

        self.stok_durumu_cerceve = tk.LabelFrame(self.ana_pencere, text="Stok Durumu", bg="#FFFF00", padx=20, pady=20)
        self.stok_durumu_cerceve.place(relx=0.5, rely=0.05, relwidth=0.4, relheight=0.9)

        self.stok_durumu_guncelle()

        self.stok_guncelle_button = tk.Button(self.ana_pencere, text="Stok Güncelle", command=self.stok_guncelle_arayuzunu_ac, bg="#4169E1", fg="white", padx=10, pady=5)
        self.stok_guncelle_button.place(relx=0.75, rely=0.90, anchor="center")

    def stok_guncelle_arayuzunu_ac(self):
        stok_guncelleme_pencere = tk.Toplevel(self.ana_pencere)
        stok_guncelleme_pencere.title("Stok Güncelleme")
        stok_guncelleme_pencere.geometry("300x200")

        urun_adi_label = tk.Label(stok_guncelleme_pencere, text="Yemek Adı:", padx=5, pady=5)
        urun_adi_label.grid(row=0, column=0)

        self.urun_adi_guncelleme_entry = tk.Entry(stok_guncelleme_pencere)
        self.urun_adi_guncelleme_entry.grid(row=0, column=1, padx=5, pady=5)

        miktar_label = tk.Label(stok_guncelleme_pencere, text="Yeni Stok Miktarı:", padx=5, pady=5)
        miktar_label.grid(row=1, column=0)

        self.miktar_guncelleme_entry = tk.Entry(stok_guncelleme_pencere)
        self.miktar_guncelleme_entry.grid(row=1, column=1, padx=5, pady=5)

        fiyat_label = tk.Label(stok_guncelleme_pencere, text="Yeni Fiyat (TL):", padx=5, pady=5)
        fiyat_label.grid(row=2, column=0)

        self.fiyat_guncelleme_entry = tk.Entry(stok_guncelleme_pencere)
        self.fiyat_guncelleme_entry.grid(row=2, column=1, padx=5, pady=5)

        guncelle_button = tk.Button(stok_guncelleme_pencere, text="Güncelle", command=self.stok_guncelle_onay, bg="#4169E1", fg="white", padx=10, pady=5)
        guncelle_button.grid(row=3, columnspan=2, pady=10)

    def stok_guncelle_onay(self):
        urun_adi = self.urun_adi_guncelleme_entry.get()
        yeni_stok_miktari = int(self.miktar_guncelleme_entry.get())
        yeni_fiyat = float(self.fiyat_guncelleme_entry.get())

        if self.stok_sistemi.stok_guncelle(urun_adi, yeni_stok_miktari, yeni_fiyat):
            messagebox.showinfo("Başarılı", f"'{urun_adi}' yemeği başarıyla güncellendi.")
            self.stok_durumu_guncelle()
        else:
            messagebox.showerror("Hata", f"'{urun_adi}' yemeği bulunamadı.")

    def stok_durumu_guncelle(self):
        for widget in self.stok_durumu_cerceve.winfo_children():
            widget.destroy()

        self.stok_durumu_label = tk.Label(self.stok_durumu_cerceve, text="Stok Durumu", bg="#FFFF00", padx=5, pady=5, font=("Helvetica", 16))
        self.stok_durumu_label.pack()

        for i, urun in enumerate(self.stok_sistemi.urunler):
            urun_label = tk.Label(self.stok_durumu_cerceve, text=f"{urun.urun_adi}: Stok: {urun.stok_miktari}, Fiyat: {urun.fiyat} TL", bg="#FFFF00", padx=5, pady=5)
            urun_label.pack(anchor="w")

    def siparis_olustur(self):
        urun_adi = self.urun_adi_entry.get()
        adres = self.adres_entry.get()
        musteri_ad = self.musteri_ad_entry.get()
        siparis_no = self.siparis_no_entry.get()

        urun_var = False
        for urun in self.stok_sistemi.urunler:
            if urun.urun_adi.lower() == urun_adi.lower():
                urun_var = True
                # Stoktan düşme işlemi
                urun.stok_miktari -= 1
                # Sipariş oluşturma
                siparis = Siparis(siparis_no, musteri_ad, adres)
                siparis.siparis_detay_ekle(urun_adi)
                self.mevcut_siparisler.append(siparis)
                messagebox.showinfo("Başarılı", f"'{siparis_no}' numaralı sipariş başarıyla oluşturuldu.")
                # Stok durumunu güncelle
                self.stok_durumu_guncelle()
                break

        if not urun_var:
            messagebox.showerror("Hata", f"'{urun_adi}' yemeği bulunamadı.")


    def mevcut_siparisleri_goster(self):
        self.mevcut_siparisler_pencere = tk.Toplevel(self.ana_pencere)
        self.mevcut_siparisler_pencere.title("Mevcut Siparişler")
        self.mevcut_siparisler_pencere.geometry("600x400")

        mevcut_siparisler_listbox = tk.Listbox(self.mevcut_siparisler_pencere, width=80, height=20)
        mevcut_siparisler_listbox.pack(padx=10, pady=10)

        mevcut_siparisler_listbox.bind("<Double-1>", lambda event: self.siparis_detaylarini_goster(mevcut_siparisler_listbox.get(tk.ACTIVE)))

        for siparis in self.mevcut_siparisler:
            mevcut_siparisler_listbox.insert(tk.END, f"Sipariş No: {siparis.siparis_no}, Müşteri Adı: {siparis.musteri_ad}, Adres: {siparis.adres}, Sipariş Detayı: {', '.join(siparis.siparis_detaylari)}")

    def siparis_detaylarini_goster(self, siparis_bilgisi):
        siparis_no = siparis_bilgisi.split(",")[0].split(":")[1].strip()
        for siparis in self.mevcut_siparisler:
            if siparis.siparis_no == siparis_no:
                siparis_detaylari_pencere = tk.Toplevel(self.mevcut_siparisler_pencere)
                siparis_detaylari_pencere.title("Sipariş Detayları")
                siparis_detaylari_pencere.geometry("400x300")

                siparis_detaylari_label = tk.Label(siparis_detaylari_pencere, text=f"{', '.join([f'{urun} ({fiyat} TL)' for urun, fiyat in siparis.siparis_detaylari])}", padx=10, pady=10)
                siparis_detaylari_label.pack()

                siparis_sil_button = tk.Button(siparis_detaylari_pencere, text="Siparişi İptal Et", command=lambda: self.siparisi_iptal_et(siparis_no), bg="red", fg="white", padx=10, pady=5)
                siparis_sil_button.pack(pady=10)

                break

    def siparisi_iptal_et(self, siparis_no):
        for i, siparis in enumerate(self.mevcut_siparisler):
            if siparis.siparis_no == siparis_no:
                del self.mevcut_siparisler[i]
                messagebox.showinfo("Başarılı", f"'{siparis_no}' numaralı sipariş başarıyla iptal edildi.")
                self.mevcut_siparisler_pencere.destroy()
                break

stok_sistemi = Stok()
stok_sistemi.urun_ekle(Urun("Kebap", 10, 20))
stok_sistemi.urun_ekle(Urun("Pizza", 20, 30))
stok_sistemi.urun_ekle(Urun("Makarna", 15, 15))
stok_sistemi.urun_ekle(Urun("Salata", 5, 10))

arayuz = SiparisArayuzu(stok_sistemi)
arayuz.ana_pencere.mainloop()
```
