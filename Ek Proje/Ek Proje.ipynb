```python
import tkinter as tk
from tkinter import messagebox, simpledialog
from datetime import datetime

class Oda:
    def __init__(self, oda_tipi, oda_numarasi, detaylar, fiyat):
        self.oda_tipi = oda_tipi
        self.oda_numarasi = oda_numarasi
        self.detaylar = detaylar
        self.fiyat = fiyat
        self.kiralanma_durumu = False
        self.giris_tarihi = None
        self.cikis_tarihi = None

    def kira(self, giris_tarihi, cikis_tarihi, fiyat):
        self.kiralanma_durumu = True
        self.giris_tarihi = giris_tarihi
        self.cikis_tarihi = cikis_tarihi
        self.fiyat = fiyat

class Otel:
    def __init__(self):
        self.odalar = []

    def oda_ekle(self, oda):
        self.odalar.append(oda)

class OtelArayuzu:
    def __init__(self, otel):
        self.otel = otel

        self.ana_pencere = tk.Tk()
        self.ana_pencere.title("Otel Yönetim Sistemi")
        self.ana_pencere.geometry("800x600")

        self.sol_panel = tk.Frame(self.ana_pencere, bg="#F08080")
        self.sol_panel.place(relx=0.05, rely=0.05, relwidth=0.4, relheight=0.9)

        self.sag_panel = tk.Frame(self.ana_pencere, bg="#87CEEB")
        self.sag_panel.place(relx=0.55, rely=0.05, relwidth=0.4, relheight=0.9)

        self.odalar_label = tk.Label(self.sol_panel, text="Kiralanan Odalar", bg="#F08080", padx=5, pady=5, font=("Helvetica", 16))
        self.odalar_label.pack()

        self.kiralanabilir_label = tk.Label(self.sag_panel, text="Kiralanabilir Odalar", bg="#87CEEB", padx=5, pady=5, font=("Helvetica", 16))
        self.kiralanabilir_label.pack()

        self.odalari_listele()

        self.oda_ekle_buton = tk.Button(self.sag_panel, text="Oda Ekle", bg="#32CD32", padx=10, pady=5, command=self.oda_ekle_arayuzu_ac)
        self.oda_ekle_buton.pack(side=tk.BOTTOM, padx=5, pady=10)

    def odalari_listele(self):
        renkler = {"VIP": "#FFD700", "Çift Kişilik": "#98FB98", "Tek Kişilik": "#ADD8E6"}

        for oda in self.otel.odalar:
            if not oda.kiralanma_durumu:
                self.oda_butonu_olustur(self.sag_panel, oda, renkler[oda.oda_tipi])
            else:
                self.oda_butonu_olustur(self.sol_panel, oda, "#FFFFFF")

    def oda_butonu_olustur(self, panel, oda, arka_plan_rengi):
        oda_butonu = tk.Button(panel, text=f"{oda.oda_tipi} - {oda.oda_numarasi}", bg=arka_plan_rengi, padx=10, pady=5, command=lambda: self.detay_goster(oda))
        oda_butonu.pack(side=tk.TOP, padx=5, pady=5)

    def oda_ekle_arayuzu_ac(self):
        oda_ekleme_penceresi = tk.Toplevel(self.ana_pencere)
        oda_ekleme_penceresi.title("Oda Ekle")
        oda_ekleme_penceresi.geometry("300x200")

        oda_tipi_label = tk.Label(oda_ekleme_penceresi, text="Oda Türü:", padx=5, pady=5)
        oda_tipi_label.grid(row=0, column=0)

        oda_tipi_var = tk.StringVar()
        oda_tipi_var.set("VIP")
        oda_tipi_combobox = tk.OptionMenu(oda_ekleme_penceresi, oda_tipi_var, "VIP", "Çift Kişilik", "Tek Kişilik")
        oda_tipi_combobox.grid(row=0, column=1, padx=5, pady=5)

        oda_numarasi_label = tk.Label(oda_ekleme_penceresi, text="Oda Numarası:", padx=5, pady=5)
        oda_numarasi_label.grid(row=1, column=0)

        oda_numarasi_entry = tk.Entry(oda_ekleme_penceresi)
        oda_numarasi_entry.grid(row=1, column=1, padx=5, pady=5)

        detaylar_label = tk.Label(oda_ekleme_penceresi, text="Detaylar:", padx=5, pady=5)
        detaylar_label.grid(row=2, column=0)

        detaylar_entry = tk.Entry(oda_ekleme_penceresi)
        detaylar_entry.grid(row=2, column=1, padx=5, pady=5)

        fiyat_label = tk.Label(oda_ekleme_penceresi, text="Fiyat (TL):", padx=5, pady=5)
        fiyat_label.grid(row=3, column=0)

        fiyat_entry = tk.Entry(oda_ekleme_penceresi)
        fiyat_entry.grid(row=3, column=1, padx=5, pady=5)

        ekle_button = tk.Button(oda_ekleme_penceresi, text="Ekle", command=lambda: self.oda_ekle(oda_tipi_var.get(), oda_numarasi_entry.get(), detaylar_entry.get(), fiyat_entry.get(), oda_ekleme_penceresi), bg="#32CD32", fg="white", padx=10, pady=5)
        ekle_button.grid(row=4, columnspan=2, pady=10)

    def oda_ekle(self, oda_tipi, oda_numarasi, detaylar, fiyat, pencere):
        if oda_tipi and oda_numarasi and detaylar and fiyat:
            if oda_numarasi.isdigit():
                if not any(oda.oda_numarasi == oda_numarasi for oda in self.otel.odalar):
                    self.otel.oda_ekle(Oda(oda_tipi, oda_numarasi, detaylar, fiyat))
                    renkler = {"VIP": "#FFD700", "Çift Kişilik": "#98FB98", "Tek Kişilik": "#ADD8E6"}
                    self.oda_butonu_olustur(self.sag_panel, Oda(oda_tipi, oda_numarasi, detaylar, fiyat), renkler[oda_tipi])
                    messagebox.showinfo("Başarılı", "Oda başarıyla eklendi.")
                    pencere.destroy()
                else:
                    messagebox.showerror("Hata", "Bu oda numarası zaten kullanımda.")
            else:
                messagebox.showerror("Hata", "Oda numarası sadece sayılardan oluşmalıdır.")
        else:
            messagebox.showerror("Hata", "Lütfen tüm alanları doldurun.")

    def detay_goster(self, oda):
        detaylar = f"Oda Numarası: {oda.oda_numarasi}\nOda Türü: {oda.oda_tipi}\n\nDetaylar:\n{oda.detaylar}\n\n"
        if oda.kiralanma_durumu:
            kalinan_gun_sayisi = (oda.cikis_tarihi - oda.giris_tarihi).days
            toplam_fiyat = kalinan_gun_sayisi * oda.fiyat
            detaylar += f"Günlük Fiyat: {oda.fiyat/kalinan_gun_sayisi} TL\nToplam Fiyat: {toplam_fiyat/kalinan_gun_sayisi} TL"
        else:
            detaylar += f"Fiyat: {oda.fiyat} TL"
        detay_penceresi = tk.Toplevel(self.ana_pencere)
        detay_penceresi.title(f"{oda.oda_tipi} - {oda.oda_numarasi} Detayları")
        detay_penceresi.geometry("300x250")
        detay_label = tk.Label(detay_penceresi, text=detaylar, padx=10, pady=10)
        detay_label.pack()
        if not oda.kiralanma_durumu:
            kirala_button = tk.Button(detay_penceresi, text="Kirala", command=lambda: self.kirala(oda), bg="#32CD32", fg="white", padx=10, pady=5)
            kirala_button.pack()

    def kirala(self, oda):
        giris_tarihi = simpledialog.askstring("Giriş Tarihi", "Giriş tarihini girin (GG/AA/YYYY):")
        cikis_tarihi = simpledialog.askstring("Çıkış Tarihi", "Çıkış tarihini girin (GG/AA/YYYY):")

        renkler = {"VIP": "#FFD700", "Çift Kişilik": "#98FB98", "Tek Kişilik": "#ADD8E6"}

        if giris_tarihi and cikis_tarihi:
            try:
                giris_tarihi = datetime.strptime(giris_tarihi, "%d/%m/%Y")
                cikis_tarihi = datetime.strptime(cikis_tarihi, "%d/%m/%Y")
            except ValueError:
                messagebox.showerror("Hata", "Lütfen tarihleri GG/AA/YYYY formatında girin.")
                return
            if giris_tarihi < datetime.now() or cikis_tarihi < datetime.now():
                messagebox.showerror("Hata", "Geçmiş tarihler için kiralama yapılamaz.")
            elif giris_tarihi >= cikis_tarihi:
                messagebox.showerror("Hata", "Çıkış tarihi, giriş tarihinden sonra olmalıdır.")
            else:
                kalinan_gun_sayisi = (cikis_tarihi - giris_tarihi).days
                fiyat = kalinan_gun_sayisi * oda.fiyat
                oda.kira(giris_tarihi, cikis_tarihi, fiyat)
                self.oda_butonu_olustur(self.sol_panel, oda, renkler[oda.oda_tipi])
                messagebox.showinfo("Başarılı", "Oda kiralama işlemi başarıyla gerçekleştirildi.")

if __name__ == "__main__":
    otel = Otel()
    otel.oda_ekle(Oda("VIP", "101", "Özel jakuzili", 500))
    otel.oda_ekle(Oda("VIP", "102", "Manzara gören", 700))
    otel.oda_ekle(Oda("Çift Kişilik", "201", "Kral yataklı", 300))
    otel.oda_ekle(Oda("Çift Kişilik", "202", "Balkonlu", 400))
    otel.oda_ekle(Oda("Tek Kişilik", "301", "Standart", 200))
    otel.oda_ekle(Oda("Tek Kişilik", "302", "Deniz manzaralı", 250))

    arayuz = OtelArayuzu(otel)
    arayuz.ana_pencere.mainloop()
```
