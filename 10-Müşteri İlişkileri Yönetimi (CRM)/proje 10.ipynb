```python
import tkinter as tk

class Müşteri:
    def __init__(self, ad, iletişim):
        self.ad = ad
        self.iletişim = iletişim
        self.siparişler = []

    def sipariş_ekle(self, ürün, miktar):
        self.siparişler.append((ürün, miktar))

# Örnek olarak bir müşteri veritabanı oluşturalım
müşteri_veritabanı = {}

# Stok miktarı
stok_miktarı = 10000

def müşteri_ekle():
    ad = entry_ad.get()
    iletişim = entry_iletişim.get()
    yeni_müşteri = Müşteri(ad, iletişim)
    müşteri_veritabanı[ad] = yeni_müşteri
    label_durum.config(text="Müşteri eklendi: " + ad)

def sipariş_ekle():
    global stok_miktarı
    ad = entry_ad.get()
    ürün = entry_ürün.get()
    miktar = int(entry_miktar.get())

    if ad in müşteri_veritabanı:
        müşteri = müşteri_veritabanı[ad]
        if miktar <= stok_miktarı:
            müşteri.sipariş_ekle(ürün, miktar)
            stok_miktarı -= miktar
            label_durum.config(text="Sipariş eklendi: " + ürün)
        else:
            label_durum.config(text="Stokta yeterli ürün yok!")
    else:
        label_durum.config(text="Müşteri bulunamadı!")

def siparişleri_göster():
    ad = entry_ad.get()
    if ad in müşteri_veritabanı:
        müşteri = müşteri_veritabanı[ad]
        sipariş_listesi = ""
        for sipariş in müşteri.siparişler:
            ürün, miktar = sipariş
            sipariş_listesi += f"{müşteri.ad} {miktar} adet {ürün} aldı.\n"
        # Yeni pencere oluştur
        sipariş_penceresi = tk.Toplevel(root)
        sipariş_penceresi.title("Sipariş Bilgileri")
        sipariş_penceresi.geometry("300x300")
        label_siparişler = tk.Label(sipariş_penceresi, text=sipariş_listesi)
        label_siparişler.pack()
    else:
        label_siparişler.config(text="Müşteri bulunamadı!")

root = tk.Tk()
root.geometry("400x500")

label_ad = tk.Label(root, text="Müşteri Adı:")
label_ad.pack()
entry_ad = tk.Entry(root)
entry_ad.pack()

label_iletişim = tk.Label(root, text="İletişim Bilgisi:")
label_iletişim.pack()
entry_iletişim = tk.Entry(root)
entry_iletişim.pack()

button_müşteri_ekle = tk.Button(root, text="Müşteri Ekle", command=müşteri_ekle)
button_müşteri_ekle.pack()

label_ürün = tk.Label(root, text="Satılan Ürün:")
label_ürün.pack()
entry_ürün = tk.Entry(root)
entry_ürün.pack()

label_miktar = tk.Label(root, text="Miktar:")
label_miktar.pack()
entry_miktar = tk.Entry(root)
entry_miktar.pack()

button_sipariş_ekle = tk.Button(root, text="Sipariş Ekle", command=sipariş_ekle)
button_sipariş_ekle.pack()

button_sipariş_göster = tk.Button(root, text="Siparişleri Göster", command=siparişleri_göster)
button_sipariş_göster.pack()

label_durum = tk.Label(root, text="")
label_durum.pack()

root.mainloop()
```
