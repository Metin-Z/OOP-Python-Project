```python
import tkinter as tk
from tkinter import messagebox, simpledialog
import random

class Ders:
    def __init__(self, ders_adi, icerik, ogretmen_adı):
        self.ders_adi = ders_adi
        self.icerik = icerik
        self.ogretmen_adı = ogretmen_adı

class Ogrenci:
    def __init__(self, numara, isim, cinsiyet):
        self.numara = numara
        self.isim = isim
        self.cinsiyet = cinsiyet

class Takip:
    def __init__(self):
        self.ogrenci_isimleri = ["Metin", "Barış", "Han", "Ayşe", "Fatoş", "Aykut", "Şenol", "Elif", "Zeynep", "Arda"]
        self.ogrenciler = [Ogrenci(numara, random.choice(self.ogrenci_isimleri), random.choice(["Erkek", "Kız"])) for numara in range(1, 11)] 
        self.dersler = []

    def ders_ekle(self, ders):
        self.dersler.append(ders)

    def ders_atama(self, ders):
        for ogrenci in self.ogrenciler:
            devamsizlik_durumu = "Devamlı Öğrenci" if random.random() < 0.5 else "Rastgele Devamsız Öğrenci"
            messagebox.showinfo("Ders Atandı", f"Öğrenci {ogrenci.numara}: {ders.ders_adi} ({devamsizlik_durumu})")

class MateryalEklemeArayuzu:
    def __init__(self, takip_sistemi):
        self.takip_sistemi = takip_sistemi
        self.ana_pencere = tk.Tk()
        self.ana_pencere.title("Materyal Ekleme Arayüzü")

        self.dersler_frame = tk.LabelFrame(self.ana_pencere, text="Dersler", padx=20, pady=10, borderwidth=2, relief="groove", bg="#f0f0f0")
        self.dersler_frame.pack(fill="both", expand="yes", padx=20, pady=10)

        self.ders_adi_label = tk.Label(self.dersler_frame, text="Ders Adı:", bg="#f0f0f0", font=("Arial", 12))
        self.ders_adi_label.grid(row=0, column=0, sticky="w", padx=5, pady=5)
        self.ders_adi_entry = tk.Entry(self.dersler_frame, font=("Arial", 12))
        self.ders_adi_entry.grid(row=0, column=1, padx=5, pady=5)

        self.icerik_label = tk.Label(self.dersler_frame, text="İçerik:", bg="#f0f0f0", font=("Arial", 12))
        self.icerik_label.grid(row=1, column=0, sticky="w", padx=5, pady=5)
        self.icerik_entry = tk.Entry(self.dersler_frame, font=("Arial", 12))
        self.icerik_entry.grid(row=1, column=1, padx=5, pady=5)

        self.ogretmen_adı_label = tk.Label(self.dersler_frame, text="Öğretmen Adı:", bg="#f0f0f0", font=("Arial", 12))
        self.ogretmen_adı_label.grid(row=2, column=0, sticky="w", padx=5, pady=5)
        self.ogretmen_adı_entry = tk.Entry(self.dersler_frame, font=("Arial", 12))
        self.ogretmen_adı_entry.grid(row=2, column=1, padx=5, pady=5)

        self.ekle_button = tk.Button(self.dersler_frame, text="Materyal Ekle", command=self.materyal_ekle, font=("Arial", 12), bg="#4CAF50", fg="white", relief="raised", padx=10, pady=5)
        self.ekle_button.grid(row=3, columnspan=2, padx=5, pady=5)

        self.detay_button = tk.Button(self.dersler_frame, text="Detayları Göster", command=self.detaylari_goster, font=("Arial", 12), bg="#008CBA", fg="white", relief="raised", padx=10, pady=5)
        self.detay_button.grid(row=4, columnspan=2, padx=5, pady=5)

        self.ogrenciler_button = tk.Button(self.ana_pencere, text="Öğrenciler", command=self.ogrencileri_goster, font=("Arial", 12), bg="#FFD700", relief="raised", padx=10, pady=5)
        self.ogrenciler_button.pack(pady=5)

    def materyal_ekle(self):
        ders_adi = self.ders_adi_entry.get()
        icerik = self.icerik_entry.get()
        ogretmen_adı = self.ogretmen_adı_entry.get()

        if ders_adi and icerik and ogretmen_adı:
            ders = Ders(ders_adi, icerik, ogretmen_adı)
            self.takip_sistemi.ders_ekle(ders)
            messagebox.showinfo("Başarılı", "Materyal başarıyla eklendi.")
            self.ders_adi_entry.delete(0, tk.END)
            self.icerik_entry.delete(0, tk.END)
            self.ogretmen_adı_entry.delete(0, tk.END)
        else:
            messagebox.showerror("Hata", "Lütfen tüm alanları doldurun.")

    def detaylari_goster(self):
        if not self.takip_sistemi.dersler:
            messagebox.showinfo("Bilgi", "Henüz materyal eklenmemiş.")
            return

        detaylar_pencere = tk.Toplevel(self.ana_pencere)
        detaylar_pencere.title("Materyal Detayları")

        ders_adi_secim_label = tk.Label(detaylar_pencere, text="Ders Seçin:", font=("Arial", 12))
        ders_adi_secim_label.pack()

        ders_adi_listebox = tk.Listbox(detaylar_pencere, selectmode=tk.SINGLE, font=("Arial", 12), height=5)
        for ders in self.takip_sistemi.dersler:
            ders_adi_listebox.insert(tk.END, ders.ders_adi)
        ders_adi_listebox.pack(pady=10)

        detay_label = tk.Label(detaylar_pencere, font=("Arial", 12), wraplength=400)
        detay_label.pack(padx=20, pady=10)

        def detaylari_goster():
            selected_index = ders_adi_listebox.curselection()
            if selected_index:
                selected_ders = self.takip_sistemi.dersler[selected_index[0]]
                detaylar = f"Ders Adı: {selected_ders.ders_adi}\n\nİçerik:\n{selected_ders.icerik}\n\nÖğretmen Adı: {selected_ders.ogretmen_adı}"
                detay_label.config(text=detaylar)
                
                ogrenci_detaylar = "\n\nÖğrenci Detayları:\n"
                for ogrenci in self.takip_sistemi.ogrenciler:
                    devamsizlik_durumu = "Devamlı Öğrenci" if random.random() < 0.5 else "Devamsız Öğrenci"
                    ogrenci_detaylar += f"Öğrenci {ogrenci.numara}: {ogrenci.isim} - {devamsizlik_durumu}\n"
                detay_label.config(text=detaylar + ogrenci_detaylar + "\n" + "-"*50)
                
                def soru_sor():
                    soru = simpledialog.askstring("Soru Sorma", f"Lütfen {selected_ders.ogretmen_adı} hocaya bir soru yazın:")
                    if soru:
                        messagebox.showinfo("Başarılı", f"Sorunuz {selected_ders.ogretmen_adı} hocaya gönderildi.")
                    else:
                        messagebox.showinfo("Bilgi", "Soru gönderilmedi.")
                
                soru_sor_button = tk.Button(detaylar_pencere, text="Soru Sor", command=soru_sor, font=("Arial", 12), bg="#4CAF50", fg="white", relief="raised", padx=10, pady=5)
                soru_sor_button.pack()
                
            else:
                messagebox.showerror("Hata", "Lütfen bir ders seçin.")

        detaylari_goster_button = tk.Button(detaylar_pencere, text="Detayları Göster", command=detaylari_goster, font=("Arial", 12), bg="#4CAF50", fg="white", relief="raised", padx=10, pady=5)
        detaylari_goster_button.pack()

    def ogrencileri_goster(self):
        ogrenciler_pencere = tk.Toplevel(self.ana_pencere)
        ogrenciler_pencere.title("Öğrenciler")

        for ogrenci in self.takip_sistemi.ogrenciler:
            ogrenci_label = tk.Label(ogrenciler_pencere, text=f"Öğrenci {ogrenci.numara}: {ogrenci.isim} - {ogrenci.cinsiyet}", font=("Arial", 12))
            ogrenci_label.pack(padx=20, pady=5)

if __name__ == "__main__":
    takip_sistemi = Takip()
    arayuz = MateryalEklemeArayuzu(takip_sistemi)
    arayuz.ana_pencere.mainloop()
```
