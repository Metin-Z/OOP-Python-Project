```python
import tkinter as tk
from tkinter import messagebox
```

```python
class AracBilgileri:
    def __init__(self, root):
        self.root = root
        self.lbl_arac_id = tk.Label(root, text="Araç ID:")
        self.lbl_arac_id.grid(row=2, column=0)
        self.entry_arac_id = tk.Entry(root)
        self.entry_arac_id.grid(row=2, column=1)
        
        self.lbl_arac_model = tk.Label(root, text="Araç Model:")
        self.lbl_arac_model.grid(row=3, column=0)
        self.entry_arac_model = tk.Entry(root)
        self.entry_arac_model.grid(row=3, column=1)
        
        self.lbl_kiralama_durumu = tk.Label(root, text="Kiralama Durumu:")
        self.lbl_kiralama_durumu.grid(row=4, column=0)
        self.entry_kiralama_durumu = tk.Entry(root)
        self.entry_kiralama_durumu.grid(row=4, column=1)
        
        self.lbl_arac_durumu = tk.Label(root, text="Araç Durumu:")
        self.lbl_arac_durumu.grid(row=5, column=0)
        self.entry_arac_durumu = tk.Entry(root)
        self.entry_arac_durumu.grid(row=5, column=1)
        
class AracKiralama:
    def __init__(self, root):
        self.root = root
        self.root.title("Araç Kiralama Uygulaması")
        self.root.geometry("1024x720")
       
        # Müşteri bilgileri
        self.lbl_ad = tk.Label(root, text="Ad :")
        self.lbl_ad.grid(row=0, column=0)
        self.entry_ad = tk.Entry(root)
        self.entry_ad.grid(row=0, column=1)
        
        self.lbl_soyad = tk.Label(root, text="Soyad :")
        self.lbl_soyad.grid(row=1, column=0)
        self.entry_soyad = tk.Entry(root)
        self.entry_soyad.grid(row=1, column=1)
        
        self.arac_bilgileri = AracBilgileri(root)
        
        # Kiralama işlemleri
        self.btn_kirala = tk.Button(root, text="Kirala", command=self.kirala)
        self.btn_kirala.grid(row=6, columnspan=2)
        
        self.btn_iptal = tk.Button(root, text="Kiralama İptal Et", command=self.kiralama_iptal_et)
        self.btn_iptal.grid(row=7, columnspan=2)
        
        self.btn_bilgi = tk.Button(root, text="Kiralama Bilgisi", command=self.kiralama_bilgisi)
        self.btn_bilgi.grid(row=8, columnspan=2)
        
    def kirala(self):
        ad = self.entry_ad.get()
        soyad = self.entry_soyad.get()
        arac_id = self.arac_bilgileri.entry_arac_id.get()
        messagebox.showinfo("Başarılı", f"{ad} {soyad} adlı müşteri {arac_id} aracını kiraladı.")
        
    def kiralama_iptal_et(self):
        kiralama_id = self.arac_bilgileri.entry_arac_id.get()  
        messagebox.showinfo("Kiralama İptal Edildi", f"{kiralama_id} numaralı kiralama iptal edildi.")
        
    def kiralama_bilgisi(self):
        kiralama_id = self.arac_bilgileri.entry_arac_id.get() 
        messagebox.showinfo("Kiralama Bilgisi", f"{kiralama_id} numaralı kiralama bilgileri görüntülendi.")
            
            
if __name__ == "__main__":
    root = tk.Tk()
    uygulama = AracKiralama(root)
    root.mainloop()

        
```
