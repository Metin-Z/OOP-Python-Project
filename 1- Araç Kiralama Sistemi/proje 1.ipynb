```python
import tkinter as tk
from tkinter import messagebox

class AracBilgileri:
    def __init__(self, root):
        self.root = root
        self.arac_listesi = {"Araç 1": {"ID": 1, "Model": "Impala", "Durum": "Müsait"},
                             "Araç 2": {"ID": 2, "Model": "Clio", "Durum": "Müsait"},
                             "Araç 3": {"ID": 3, "Model": "Leon", "Durum": "Müsait"},
                             "Araç 4": {"ID": 4, "Model": "Accord", "Durum": "Müsait"},
                             "Araç 5": {"ID": 5, "Model": "Fusion", "Durum": "Müsait"}}
        self.arac_secim = tk.StringVar(root)
        self.arac_secim.set("Araç 1")  # Default value
        self.dropdown_arac = tk.OptionMenu(root, self.arac_secim, *self.arac_listesi.keys())
        self.dropdown_arac.grid(row=6, column=1)

class AracKiralama:
    def __init__(self, root):
        self.root = root
        self.root.title("Araç Kiralama Uygulaması")
        self.root.geometry("1024x720")
        self.arac_kiralandi = False
        self.arac_bilgileri = AracBilgileri(root)
       
        # Müşteri bilgileri
        self.lbl_ad = tk.Label(root, text="Ad :")
        self.lbl_ad.grid(row=0, column=0)
        self.entry_ad = tk.Entry(root)
        self.entry_ad.grid(row=0, column=1)
        
        self.lbl_soyad = tk.Label(root, text="Soyad :")
        self.lbl_soyad.grid(row=1, column=0)
        self.entry_soyad = tk.Entry(root)
        self.entry_soyad.grid(row=1, column=1)
        
        # Kiralama işlemleri
        self.btn_kirala = tk.Button(root, text="Kirala", command=self.kirala)
        self.btn_kirala.grid(row=7, column=0)
        
        self.btn_iptal = tk.Button(root, text="Kiralama İptal Et", command=self.iptal_et)
        self.btn_iptal.grid(row=7, column=1)
        
    def kirala(self):
        ad = self.entry_ad.get()
        soyad = self.entry_soyad.get()
        arac_secim = self.arac_bilgileri.arac_secim.get()
        arac = self.arac_bilgileri.arac_listesi[arac_secim]
        if not self.arac_kiralandi and arac["Durum"] == "Müsait":
            arac["Durum"] = "Kiralandı"
            self.arac_kiralandi = True
            del self.arac_bilgileri.arac_listesi[arac_secim]  # Stoktan kiralanan aracı kaldır
            messagebox.showinfo("Başarılı", f"{ad} {soyad} adlı müşteri {arac['ID']} - {arac['Model']} aracını kiraladı.")
        elif self.arac_kiralandi:
            messagebox.showinfo("Hata", "Zaten bir araç kiraladınız!")
        else:
            messagebox.showinfo("Hata", f"{arac_secim} aracı zaten kiralandı!")
    
    def iptal_et(self):
        ad = self.entry_ad.get()
        soyad = self.entry_soyad.get()
        arac_secim = self.arac_bilgileri.arac_secim.get()
        arac = self.arac_bilgileri.arac_listesi[arac_secim]
        if self.arac_kiralandi and arac["Durum"] == "Kiralandı":
            arac["Durum"] = "Müsait"
            self.arac_kiralandi = False
            messagebox.showinfo("Başarılı", f"{ad} {soyad} adlı müşteri {arac['ID']} - {arac['Model']} aracının kiralama işlemi iptal edildi.")
        else:
            messagebox.showinfo("Hata", "Kiralama işlemi bulunamadı!")

if __name__ == "__main__":
    root = tk.Tk()
    uygulama = AracKiralama(root)
    root.mainloop()
```
