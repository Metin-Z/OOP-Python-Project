```python
import tkinter as tk

class TarihciArayuzu:
    def __init__(self):
        self.root = tk.Tk()
        self.root.title("Tarihçi - Tarihi Olaylar Veritabanı")
        self.root.geometry("1550x450")
        self.root.configure(bg="#f0f0f0")  

        self.olaylar_panel = tk.LabelFrame(self.root, text="Olaylar", padx=50, pady=20, bg="#446677", fg="white")
        self.olaylar_panel.grid(row=0, column=0, padx=15, pady=15)

        self.olay_adi_label = tk.Label(self.olaylar_panel, text="Olay Adı:", bg="#446677", fg="white")
        self.olay_adi_label.grid(row=0, column=0, padx=5, pady=5, sticky="w")
        self.olay_adi_entry = tk.Entry(self.olaylar_panel)
        self.olay_adi_entry.grid(row=0, column=1, padx=5, pady=5)

        self.tarih_label = tk.Label(self.olaylar_panel, text="Tarih:", bg="#446677", fg="white")
        self.tarih_label.grid(row=1, column=0, padx=5, pady=5, sticky="w")
        self.tarih_entry = tk.Entry(self.olaylar_panel)
        self.tarih_entry.grid(row=1, column=1, padx=5, pady=5)

        self.aciklama_label = tk.Label(self.olaylar_panel, text="Açıklama:", bg="#446677", fg="white")
        self.aciklama_label.grid(row=2, column=0, padx=5, pady=5, sticky="w")
        self.aciklama_entry = tk.Entry(self.olaylar_panel)
        self.aciklama_entry.grid(row=2, column=1, padx=5, pady=5)

        self.olay_gonder_button = tk.Button(self.olaylar_panel, text="Olay Gönder", bg="#446677", fg="white", command=self.ekle_olay)
        self.olay_gonder_button.grid(row=3, columnspan=2, pady=10)

        self.olaylar_listebox = tk.Listbox(self.olaylar_panel, height=8, width=60)
        self.olaylar_listebox.grid(row=4, columnspan=2, padx=7, pady=5)
        self.olaylar_listebox.bind('<Double-Button-1>', self.olay_detay)

        self.olay_sil_button = tk.Button(self.olaylar_panel, text="Seçiliyi Sil", bg="#446677", fg="white", command=self.olay_sil)
        self.olay_sil_button.grid(row=5, columnspan=2, pady=10)

        self.shahsiyetler_panel = tk.LabelFrame(self.root, text="Şahsiyetler", padx=50, pady=20, bg="#334455", fg="white")
        self.shahsiyetler_panel.grid(row=0, column=1, padx=15, pady=15)

        self.shahsiyet_adi_label = tk.Label(self.shahsiyetler_panel, text="Şahsiyet Adı:", bg="#334455", fg="white")
        self.shahsiyet_adi_label.grid(row=0, column=0, padx=5, pady=5, sticky="w")
        self.shahsiyet_adi_entry = tk.Entry(self.shahsiyetler_panel)
        self.shahsiyet_adi_entry.grid(row=0, column=1, padx=5, pady=5)

        self.dogum_tarihi_label = tk.Label(self.shahsiyetler_panel, text="Doğum Tarihi:", bg="#334455", fg="white")
        self.dogum_tarihi_label.grid(row=1, column=0, padx=5, pady=5, sticky="w")
        self.dogum_tarihi_entry = tk.Entry(self.shahsiyetler_panel)
        self.dogum_tarihi_entry.grid(row=1, column=1, padx=5, pady=5)

        self.olum_tarihi_label = tk.Label(self.shahsiyetler_panel, text="Ölüm Tarihi:", bg="#334455", fg="white")
        self.olum_tarihi_label.grid(row=2, column=0, padx=5, pady=5, sticky="w")
        self.olum_tarihi_entry = tk.Entry(self.shahsiyetler_panel)
        self.olum_tarihi_entry.grid(row=2, column=1, padx=5, pady=5)

        self.shahsiyet_gonder_button = tk.Button(self.shahsiyetler_panel, text="Şahsiyet Gönder", bg="#334455", fg="white", command=self.ekle_shahsiyet)
        self.shahsiyet_gonder_button.grid(row=3, columnspan=2, pady=10)

        self.shahsiyetler_listebox = tk.Listbox(self.shahsiyetler_panel, height=8, width=60)
        self.shahsiyetler_listebox.grid(row=4, columnspan=2, padx=7, pady=5)
        self.shahsiyetler_listebox.bind('<Double-Button-1>', self.shahsiyet_detay)

        self.shahsiyet_sil_button = tk.Button(self.shahsiyetler_panel, text="Seçiliyi Sil", bg="#334455", fg="white", command=self.shahsiyet_sil)
        self.shahsiyet_sil_button.grid(row=5, columnspan=2, pady=10)

        self.donemler_panel = tk.LabelFrame(self.root, text="Dönemler", padx=50, pady=20, bg="#223344", fg="white")
        self.donemler_panel.grid(row=0, column=2, padx=15, pady=15)

        self.donem_adi_label = tk.Label(self.donemler_panel, text="Dönem Adı:", bg="#223344", fg="white")
        self.donem_adi_label.grid(row=0, column=0, padx=5, pady=5, sticky="w")
        self.donem_adi_entry = tk.Entry(self.donemler_panel)
        self.donem_adi_entry.grid(row=0, column=1, padx=5, pady=5)

        self.baslangic_tarihi_label = tk.Label(self.donemler_panel, text="Başlangıç Tarihi:", bg="#223344", fg="white")
        self.baslangic_tarihi_label.grid(row=1, column=0, padx=5, pady=5, sticky="w")
        self.baslangic_tarihi_entry = tk.Entry(self.donemler_panel)
        self.baslangic_tarihi_entry.grid(row=1, column=1, padx=5, pady=5)

        self.bitis_tarihi_label = tk.Label(self.donemler_panel, text="Bitiş Tarihi:", bg="#223344", fg="white")
        self.bitis_tarihi_label.grid(row=2, column=0, padx=5, pady=5, sticky="w")
        self.bitis_tarihi_entry = tk.Entry(self.donemler_panel)
        self.bitis_tarihi_entry.grid(row=2, column=1, padx=7, pady=5)

        self.donem_gonder_button = tk.Button(self.donemler_panel, text="Dönem Gönder", bg="#223344", fg="white", command=self.ekle_donem)
        self.donem_gonder_button.grid(row=3, columnspan=2, pady=10)

        self.donemler_listebox = tk.Listbox(self.donemler_panel, height=8, width=60)
        self.donemler_listebox.grid(row=4, columnspan=2, padx=7, pady=5)
        self.donemler_listebox.bind('<Double-Button-1>', self.donem_detay)

        self.donem_sil_button = tk.Button(self.donemler_panel, text="Seçiliyi Sil", bg="#223344", fg="white", command=self.donem_sil)
        self.donem_sil_button.grid(row=5, columnspan=2, pady=10)

        self.root.mainloop()

    def ekle_olay(self):
        olay_adi = self.olay_adi_entry.get()
        tarih = self.tarih_entry.get()
        aciklama = self.aciklama_entry.get()

        if olay_adi and tarih and aciklama:
            self.olaylar_listebox.insert(tk.END, f"Olay Adı: {olay_adi}, Tarih: {tarih}, Açıklama: {aciklama}")
            self.olay_adi_entry.delete(0, tk.END)
            self.tarih_entry.delete(0, tk.END)
            self.aciklama_entry.delete(0, tk.END)

    def ekle_shahsiyet(self):
        shahsiyet_adi = self.shahsiyet_adi_entry.get()
        dogum_tarihi = self.dogum_tarihi_entry.get()
        olum_tarihi = self.olum_tarihi_entry.get()

        if shahsiyet_adi and dogum_tarihi and olum_tarihi:
            self.shahsiyetler_listebox.insert(tk.END, f"Şahsiyet Adı: {shahsiyet_adi}, Doğum Tarihi: {dogum_tarihi}, Ölüm Tarihi: {olum_tarihi}")
            self.shahsiyet_adi_entry.delete(0, tk.END)
            self.dogum_tarihi_entry.delete(0, tk.END)
            self.olum_tarihi_entry.delete(0, tk.END)

    def ekle_donem(self):
        donem_adi = self.donem_adi_entry.get()
        baslangic_tarihi = self.baslangic_tarihi_entry.get()
        bitis_tarihi = self.bitis_tarihi_entry.get()

        if donem_adi and baslangic_tarihi and bitis_tarihi:
            self.donemler_listebox.insert(tk.END, f"Dönem Adı: {donem_adi}, Başlangıç Tarihi: {baslangic_tarihi}, Bitiş Tarihi: {bitis_tarihi}")
            self.donem_adi_entry.delete(0, tk.END)
            self.baslangic_tarihi_entry.delete(0, tk.END)
            self.bitis_tarihi_entry.delete(0, tk.END)

    def olay_detay(self, event):
        index = self.olaylar_listebox.curselection()[0]
        selected_item = self.olaylar_listebox.get(index)
        self.detay_penceresi(selected_item)

    def shahsiyet_detay(self, event):
        index = self.shahsiyetler_listebox.curselection()[0]
        selected_item = self.shahsiyetler_listebox.get(index)
        self.detay_penceresi(selected_item)

    def donem_detay(self, event):
        index = self.donemler_listebox.curselection()[0]
        selected_item = self.donemler_listebox.get(index)
        self.detay_penceresi(selected_item)

    def detay_penceresi(self, selected_item):
        detay_pencere = tk.Toplevel(self.root)
        detay_pencere.title("Detaylar")
        detay_pencere.geometry("400x200")

        detay_etiket = tk.Label(detay_pencere, text=selected_item)
        detay_etiket.pack(padx=20, pady=20)

    def olay_sil(self):
        selected_index = self.olaylar_listebox.curselection()
        if selected_index:
            self.olaylar_listebox.delete(selected_index)

    def shahsiyet_sil(self):
        selected_index = self.shahsiyetler_listebox.curselection()
        if selected_index:
            self.shahsiyetler_listebox.delete(selected_index)

    def donem_sil(self):
        selected_index = self.donemler_listebox.curselection()
        if selected_index:
            self.donemler_listebox.delete(selected_index)

if __name__ == "__main__":
    arayuz = TarihciArayuzu()
```
