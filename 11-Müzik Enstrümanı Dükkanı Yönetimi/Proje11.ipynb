```python
import tkinter as tk
from tkinter import messagebox
import random
from tkinter import ttk

class Enstruman:
    def __init__(self, enstruman_adi, marka, stok_miktari, fiyat):
        self.enstruman_adi = enstruman_adi
        self.marka = marka
        self.stok_miktari = stok_miktari
        self.fiyat = fiyat

class Dukkan:
    def __init__(self):
        self.enstrumanlar = []

    def enstruman_ekle(self, enstruman):
        self.enstrumanlar.append(enstruman)

    def stok_guncelle(self, enstruman_adi, yeni_stok_miktari):
        for enstruman in self.enstrumanlar:
            if enstruman.enstruman_adi.lower() == enstruman_adi.lower():
                enstruman.stok_miktari = yeni_stok_miktari
                return True
        return False

    def stok_durumu_goster(self):
        stok_durumu = ""
        for enstruman in self.enstrumanlar:
            stok_durumu += f"{enstruman.enstruman_adi} ({enstruman.marka}): Miktar: {enstruman.stok_miktari}, Fiyat: {enstruman.fiyat} TL\n"
        messagebox.showinfo("Stok Durumu", stok_durumu)

class Satis:
    def __init__(self, satis_no, enstruman_adi, marka, miktar, toplam_tutar):
        self.satis_no = satis_no
        self.enstruman_adi = enstruman_adi
        self.marka = marka
        self.miktar = miktar
        self.toplam_tutar = toplam_tutar

class DukkanArayuzu:
    def __init__(self, dukkan_sistemi):
        self.dukkan_sistemi = dukkan_sistemi
        self.mevcut_satislar = []

        self.ana_pencere = tk.Tk()
        self.ana_pencere.title("Müzik Enstrümanı Dükkanı Yönetim Sistemi")
        self.ana_pencere.geometry("800x600")

        self.ana_pencere.configure(bg="#778899")

        self.satis_secim_cerceve = tk.LabelFrame(self.ana_pencere, text="Satış Yap", bg="#B0C4DE", padx=20, pady=20)
        self.satis_secim_cerceve.place(relx=0.05, rely=0.05, relwidth=0.4, relheight=0.35)

        self.urun_adi_label = tk.Label(self.satis_secim_cerceve, text="Enstrüman Adı:", bg="#B0C4DE", padx=5, pady=5)
        self.urun_adi_label.grid(row=0, column=0, sticky="w")

        self.urun_adi_entry = tk.Entry(self.satis_secim_cerceve)
        self.urun_adi_entry.grid(row=0, column=1, padx=5, pady=5)

        self.marka_label = tk.Label(self.satis_secim_cerceve, text="Marka:", bg="#B0C4DE", padx=5, pady=5)
        self.marka_label.grid(row=1, column=0, sticky="w")

        self.marka_entry = tk.Entry(self.satis_secim_cerceve)
        self.marka_entry.grid(row=1, column=1, padx=5, pady=5)

        self.miktar_label = tk.Label(self.satis_secim_cerceve, text="Miktar:", bg="#B0C4DE", padx=5, pady=5)
        self.miktar_label.grid(row=2, column=0, sticky="w")

        self.miktar_entry = tk.Entry(self.satis_secim_cerceve)
        self.miktar_entry.grid(row=2, column=1, padx=5, pady=5)

        self.satis_no_label = tk.Label(self.satis_secim_cerceve, text="Satış No:", bg="#B0C4DE", padx=5, pady=5)
        self.satis_no_label.grid(row=3, column=0, sticky="w")

        self.satis_no_entry = tk.Entry(self.satis_secim_cerceve)
        self.satis_no_entry.grid(row=3, column=1, padx=5, pady=5)

        self.ekle_button = tk.Button(self.satis_secim_cerceve, text="Satış Yap", command=self.satis_yap, bg="#4682B4", fg="white", padx=10, pady=5)
        self.ekle_button.grid(row=4, columnspan=2, pady=10)

        self.mevcut_satislar_button = tk.Button(self.ana_pencere, text="Mevcut Satışları Göster", command=self.mevcut_satislari_goster, bg="#B0C4DE", padx=10, pady=5)
        self.mevcut_satislar_button.place(relx=0.05, rely=0.45)

        self.stok_durumu_cerceve = tk.LabelFrame(self.ana_pencere, text="Stok Durumu", bg="#B0C4DE", padx=20, pady=20)
        self.stok_durumu_cerceve.place(relx=0.5, rely=0.05, relwidth=0.4, relheight=0.9)

        self.stok_durumu_guncelle()

    def stok_durumu_guncelle(self):
        for widget in self.stok_durumu_cerceve.winfo_children():
            widget.destroy()

        self.stok_durumu_label = tk.Label(self.stok_durumu_cerceve, text="Stok Durumu", bg="#B0C4DE", padx=5, pady=5, font=("Helvetica", 16))
        self.stok_durumu_label.pack()

        for i, enstruman in enumerate(self.dukkan_sistemi.enstrumanlar):
            enstruman_label = tk.Label(self.stok_durumu_cerceve, text=f"{enstruman.enstruman_adi} ({enstruman.marka}): Miktar: {enstruman.stok_miktari}, Fiyat: {enstruman.fiyat} TL", bg="#B0C4DE", padx=5, pady=5)
            enstruman_label.pack(anchor="w")

    def satis_yap(self):
        enstruman_adi = self.urun_adi_entry.get()
        marka = self.marka_entry.get()
        miktar = int(self.miktar_entry.get())
        satis_no = self.satis_no_entry.get()

        enstruman_var = False
        for enstruman in self.dukkan_sistemi.enstrumanlar:
            if enstruman.enstruman_adi.lower() == enstruman_adi.lower() and enstruman.marka.lower() == marka.lower():
                enstruman_var = True
                if enstruman.stok_miktari >= miktar:
                    toplam_tutar = miktar * enstruman.fiyat
                    destek_numarasi = random.randint(1, 1000)
                    satis = Satis(satis_no, enstruman_adi, marka, miktar, toplam_tutar)
                    self.mevcut_satislar.append(satis)
                    messagebox.showinfo("Başarılı", f"'{satis_no}' numaralı satış başarıyla gerçekleştirildi. Destek Numarası: {destek_numarasi}")
                    enstruman.stok_miktari -= miktar
                    self.stok_durumu_guncelle()
                else:
                    messagebox.showerror("Hata", f"Stokta yeterli '{enstruman_adi}' ({marka}) bulunmamaktadır. Stoktaki miktar: {enstruman.stok_miktari}")
                break

        if not enstruman_var:
            messagebox.showerror("Hata", f"'{enstruman_adi}' ({marka}) enstrümanı bulunamadı.")

    def iptal_et(self):
        selection = self.mevcut_satislar_tree.selection()
        if selection:
            for sel in selection:
                item = self.mevcut_satislar_tree.item(sel)
                satis_no = item['values'][0]
                for satis in self.mevcut_satislar:
                    if satis.satis_no == satis_no:
                        for enstruman in self.dukkan_sistemi.enstrumanlar:
                            if enstruman.enstruman_adi.lower() == satis.enstruman_adi.lower() and enstruman.marka.lower() == satis.marka.lower():
                                enstruman.stok_miktari += satis.miktar
                                break
                        self.mevcut_satislar.remove(satis)
                        self.mevcut_satislar_tree.delete(sel)
                        messagebox.showinfo("Başarılı", f"'{satis_no}' numaralı satış başarıyla iptal edildi.")
                        self.stok_durumu_guncelle()
                        break
        else:
            messagebox.showerror("Hata", "Lütfen iptal etmek istediğiniz satışı seçin.")

    def mevcut_satislari_goster(self):
        self.mevcut_satislar_pencere = tk.Toplevel(self.ana_pencere)
        self.mevcut_satislar_pencere.title("Mevcut Satışlar")
        self.mevcut_satislar_pencere.geometry("600x400")

        self.mevcut_satislar_tree = ttk.Treeview(self.mevcut_satislar_pencere, columns=("Satış No", "Enstrüman", "Miktar", "Toplam Tutar"), show="headings")
        self.mevcut_satislar_tree.heading("Satış No", text="Satış No")
        self.mevcut_satislar_tree.heading("Enstrüman", text="Enstrüman")
        self.mevcut_satislar_tree.heading("Miktar", text="Miktar")
        self.mevcut_satislar_tree.heading("Toplam Tutar", text="Toplam Tutar")

        self.mevcut_satislar_tree.pack(fill="both", expand=True)

        for satis in self.mevcut_satislar:
            self.mevcut_satislar_tree.insert("", "end", values=(satis.satis_no, f"{satis.enstruman_adi} ({satis.marka})", satis.miktar, satis.toplam_tutar))

        self.iptal_button = tk.Button(self.mevcut_satislar_pencere, text="İptal Et", command=self.iptal_et, bg="#DC143C", fg="white", padx=10, pady=5)
        self.iptal_button.pack()

if __name__ == "__main__":
    dukkan_sistemi = Dukkan()
    dukkan_sistemi.enstruman_ekle(Enstruman("Gitar", "Fender", 20, 500))
    dukkan_sistemi.enstruman_ekle(Enstruman("Gitar", "Gibson", 15, 700))
    dukkan_sistemi.enstruman_ekle(Enstruman("Piyano", "Yamaha", 10, 2000))
    dukkan_sistemi.enstruman_ekle(Enstruman("Piyano", "Kawai", 8, 2200))
    dukkan_sistemi.enstruman_ekle(Enstruman("Keman", "Stradivarius", 15, 300))
    dukkan_sistemi.enstruman_ekle(Enstruman("Keman", "Yamaha", 10, 350))
    dukkan_sistemi.enstruman_ekle(Enstruman("Davul", "Pearl", 12, 700))
    dukkan_sistemi.enstruman_ekle(Enstruman("Davul", "Ludwig", 10, 750))
    dukkan_sistemi.enstruman_ekle(Enstruman("Flüt", "Yamaha", 18, 150))
    dukkan_sistemi.enstruman_ekle(Enstruman("Flüt", "Pearl", 15, 170))

    arayuz = DukkanArayuzu(dukkan_sistemi)
    arayuz.ana_pencere.mainloop()
```
