```python
import tkinter as tk
from tkinter import simpledialog

class OyunArayuzu:
    def __init__(self):
        self.root = tk.Tk()
        self.root.title("Oyun - Video Oyun Koleksiyonu Yönetimi")
        self.root.geometry("1050x700") 
        self.root.configure(bg="#f0f0f0")  
        self.secilen_oyuncu = None

        self.olaylar_panel = tk.LabelFrame(self.root, text="Olaylar", padx=50, pady=20, bg="#446677", fg="white")
        self.olaylar_panel.grid(row=0, column=0, padx=15, pady=15)

        self.oyun_ekle_panel = tk.LabelFrame(self.root, text="Oyun Ekle", padx=50, pady=20, bg="#4b4b4b", fg="white")
        self.oyun_ekle_panel.grid(row=0, column=1, padx=15, pady=15)

        button_width = 15

        self.olay_adi_label = tk.Label(self.oyun_ekle_panel, text="Oyun Adı:", bg="#4b4b4b", fg="white")
        self.olay_adi_label.grid(row=0, column=0, padx=5, pady=5, sticky="w")
        self.olay_adi_entry = tk.Entry(self.oyun_ekle_panel)
        self.olay_adi_entry.grid(row=0, column=1, padx=5, pady=5)

        self.oyun_turu_label = tk.Label(self.oyun_ekle_panel, text="Oyun Türü:", bg="#4b4b4b", fg="white")
        self.oyun_turu_label.grid(row=1, column=0, padx=5, pady=5, sticky="w")
        self.oyun_turu_entry = tk.Entry(self.oyun_ekle_panel)
        self.oyun_turu_entry.grid(row=1, column=1, padx=5, pady=5)

        self.platform_label = tk.Label(self.oyun_ekle_panel, text="Platform:", bg="#4b4b4b", fg="white")
        self.platform_label.grid(row=2, column=0, padx=5, pady=5, sticky="w")
        self.platform_entry = tk.Entry(self.oyun_ekle_panel)
        self.platform_entry.grid(row=2, column=1, padx=5, pady=5)

        self.oyun_ekle_button = tk.Button(self.oyun_ekle_panel, text="Oyun Ekle", bg="#4b4b4b", fg="white", command=self.ekle_oyun)
        self.oyun_ekle_button.grid(row=3, columnspan=2, pady=10)

        self.oyunlar_panel = tk.LabelFrame(self.root, text="Örnek Oyunlar", padx=50, pady=20, bg="#223344", fg="white")
        self.oyunlar_panel.grid(row=0, column=2, padx=15, pady=15)

        self.oyunlar_listebox = tk.Listbox(self.oyunlar_panel, height=15, width=60)
        self.oyunlar_listebox.grid(row=0, column=0, padx=7, pady=5)
        self.oyunlar_listebox.bind('<Double-Button-1>', self.oyun_detay)  # Double-click event handler

        self.favorilere_ekle_button = tk.Button(self.oyunlar_panel, text="Favorilere Ekle", bg="#446677", fg="white", command=self.favorilere_ekle, width=button_width)
        self.favorilere_ekle_button.grid(row=1, column=0, padx=5, pady=5, sticky="ew")

        self.oyunlar_liste = [
            {"ad": "GTA 5", "tur": "Aksiyon-Açık Dünya", "platform": "PC-Playstation-Xbox"},
            {"ad": "Red Dead Redemption 2", "tur": "Açık Dünya", "platform": "PC-Playstation-Xbox"},
            {"ad": "The Witcher 3: Wild Hunt", "tur": "Rol Yapma", "platform": "PC-Playstation-Xbox"},
            {"ad": "Fortnite", "tur": "Battle-Royale", "platform": "PC-Playstation-Xbox-Nintendo Switch-Mobil"},
            {"ad": "God of War", "tur": "Aksiyon-Macera", "platform": "PC-PlayStation"}
        ]

        for oyun in self.oyunlar_liste:
            self.oyunlar_listebox.insert(tk.END, f"Oyun Adı: {oyun['ad']}, Türü: {oyun['tur']}, Platformu: {oyun['platform']}")

        self.oyuncu_ekle_panel = tk.LabelFrame(self.root, text="Oyuncu Ekle", padx=50, pady=20, bg="#223344", fg="white")
        self.oyuncu_ekle_panel.grid(row=1, column=0, padx=15, pady=15, columnspan=3)

        self.oyuncu_adi_label = tk.Label(self.oyuncu_ekle_panel, text="Oyuncu Adı:", bg="#223344", fg="white")
        self.oyuncu_adi_label.grid(row=0, column=0, padx=5, pady=5, sticky="w")
        self.oyuncu_adi_entry = tk.Entry(self.oyuncu_ekle_panel)
        self.oyuncu_adi_entry.grid(row=0, column=1, padx=5, pady=5)

        self.oyuncu_ekle_button = tk.Button(self.oyuncu_ekle_panel, text="Oyuncu Ekle", bg="#223344", fg="white", command=self.ekle_oyuncu)
        self.oyuncu_ekle_button.grid(row=0, column=2, padx=5, pady=5)

        self.oyuncular_listebox = tk.Listbox(self.oyuncu_ekle_panel, height=10, width=60)
        self.oyuncular_listebox.grid(row=1, column=0, columnspan=3, padx=7, pady=5)
        self.oyuncular_listebox.bind('<Double-Button-1>', self.oyuncu_detay)  # Double-click event handler

        self.oyuncular = [
            {"ad": "Ahmet", "favoriler": ["GTA 5"], "koleksiyon": ["The Witcher 3: Wild Hunt", "Red Dead Redemption 2"], "değerlendirmeler": {"GTA 5": 10, "Red Dead Redemption 2": 8}},
            {"ad": "Mehmet", "favoriler": ["Fortnite"], "koleksiyon": ["GTA 5", "Red Dead Redemption 2"], "değerlendirmeler": {"Fortnite": 10, "Red Dead Redemption 2": 6}}
        ]

        for oyuncu in self.oyuncular:
            self.oyuncular_listebox.insert(tk.END, f"Oyuncu Adı: {oyuncu['ad']}")

        self.koleksiyona_ekle_button = tk.Button(self.oyunlar_panel, text="Koleksiyona Ekle", bg="#446677", fg="white", command=self.koleksiyona_ekle, width=button_width)
        self.koleksiyona_ekle_button.grid(row=1, column=1, padx=5, pady=5, sticky="ew")

        self.degerlendir_button = tk.Button(self.oyunlar_panel, text="Değerlendir", bg="#446677", fg="white", command=self.degerlendir, width=button_width)
        self.degerlendir_button.grid(row=0, column=1, padx=5, pady=5, sticky="ew")

        self.root.mainloop()

    def ekle_oyun(self):
        oyun_adi = self.olay_adi_entry.get()
        oyun_turu = self.oyun_turu_entry.get()
        platform = self.platform_entry.get()

        if oyun_adi and oyun_turu and platform:
            self.oyunlar_liste.append({"ad": oyun_adi, "tur": oyun_turu, "platform": platform})
            self.oyunlar_listebox.insert(tk.END, f"Oyun Adı: {oyun_adi}, Türü: {oyun_turu}, Platformu: {platform}")
            self.olay_adi_entry.delete(0, tk.END)
            self.oyun_turu_entry.delete(0, tk.END)
            self.platform_entry.delete(0, tk.END)

        # Eğer oyuncu seçili ise favorilere ekle ve koleksiyona ekle düğmelerini etkinleştir
        if self.secilen_oyuncu:
            self.favorilere_ekle_button.config(state="normal")
            self.koleksiyona_ekle_button.config(state="normal")

    def ekle_oyuncu(self):
        oyuncu_adi = self.oyuncu_adi_entry.get()
        if oyuncu_adi:
            self.oyuncular_listebox.insert(tk.END, f"Oyuncu Adı: {oyuncu_adi}")
            self.oyuncu_adi_entry.delete(0, tk.END)

            self.oyuncular_listebox.bind('<Double-Button-1>', self.oyuncu_detay)

            self.oyuncular.append({"ad": oyuncu_adi, "favoriler": [], "koleksiyon": [], "değerlendirmeler": {}})

    def oyun_detay(self, event):
        try:
            index = self.oyunlar_listebox.curselection()[0]
            selected_item = self.oyunlar_listebox.get(index)
            self.detay_penceresi(selected_item)
        except IndexError:
            pass

    def oyuncu_detay(self, event):
        try:
            index = self.oyuncular_listebox.curselection()[0]
            selected_item = self.oyuncular_listebox.get(index)
            oyuncu_adi = selected_item.split(": ")[1]
            self.secilen_oyuncu = oyuncu_adi
            for oyuncu in self.oyuncular:
                if oyuncu['ad'] == oyuncu_adi:
                    self.detay_penceresi(f"Oyuncu Adı: {oyuncu['ad']} \n Favoriler: {', '.join(oyuncu['favoriler'])} \n Koleksiyon: {', '.join(oyuncu['koleksiyon'])}\n Değerlendirmeler: {oyuncu['değerlendirmeler']}")
        except IndexError:
            pass

    def detay_penceresi(self, selected_item):
        detay_pencere = tk.Toplevel(self.root)
        detay_pencere.title("Detaylar")
        detay_pencere.geometry("400x200")

        detay_text = tk.Text(detay_pencere, state="normal", wrap="word")
        detay_text.pack(padx=20, pady=20)

        detay_text.insert(tk.END, selected_item)

    def favorilere_ekle(self):
        selected_index = self.oyunlar_listebox.curselection()
        if selected_index and self.secilen_oyuncu:
            selected_oyun = self.oyunlar_liste[selected_index[0]]
            for oyuncu in self.oyuncular:
                if oyuncu['ad'] == self.secilen_oyuncu:
                    if selected_oyun['ad'] not in oyuncu['favoriler']:
                        oyuncu['favoriler'].append(selected_oyun['ad'])
                        self.update_oyuncular_listbox()

    def koleksiyona_ekle(self):
        selected_index = self.oyunlar_listebox.curselection()
        if selected_index and self.secilen_oyuncu:
            selected_oyun = self.oyunlar_liste[selected_index[0]]
            for oyuncu in self.oyuncular:
                if oyuncu['ad'] == self.secilen_oyuncu:
                    if selected_oyun['ad'] not in oyuncu['koleksiyon']:
                        oyuncu['koleksiyon'].append(selected_oyun['ad'])
                        self.update_oyuncular_listbox()

    def degerlendir(self):
        selected_index = self.oyunlar_listebox.curselection()
        if selected_index and self.secilen_oyuncu:
            selected_oyun = self.oyunlar_liste[selected_index[0]]
            oyuncu_index = [i for i, oyuncu in enumerate(self.oyuncular) if oyuncu['ad'] == self.secilen_oyuncu][0]
            deger = simpledialog.askinteger("Degerlendirme", f"{selected_oyun['ad']} oyununu değerlendirin (0-10 arası):")
            if deger is not None:
                self.oyuncular[oyuncu_index]['değerlendirmeler'][selected_oyun['ad']] = deger
                self.update_oyuncular_listbox()

    def update_oyuncular_listbox(self):
        self.oyuncular_listebox.delete(0, tk.END)
        for oyuncu in self.oyuncular:
            self.oyuncular_listebox.insert(tk.END, f"Oyuncu Adı: {oyuncu['ad']}")

if __name__ == "__main__":
    arayuz = OyunArayuzu()
```
