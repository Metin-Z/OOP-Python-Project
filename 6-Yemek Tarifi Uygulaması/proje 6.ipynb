```python
import tkinter as tk

class TarifUygulamasi:
    def __init__(self, ana_pencere):
        self.ana_pencere = ana_pencere
        ana_pencere.title("Tarif Uygulaması")
        ana_pencere.geometry("600x400")

        self.etiket = tk.Label(ana_pencere, text="Hoş Geldiniz!")
        self.etiket.pack()

        self.tarif_ekle_btn = tk.Button(ana_pencere, text="Tarif Ekle", command=self.tarif_ekle)
        self.tarif_ekle_btn.pack()

        self.tarif_ara_btn = tk.Button(ana_pencere, text="Tarif Ara", command=self.tarif_ara)
        self.tarif_ara_btn.pack()

        self.tarif_degerlendir_btn = tk.Button(ana_pencere, text="Tarifi Değerlendir", command=self.tarif_degerlendir)
        self.tarif_degerlendir_btn.pack()

    def tarif_ekle(self):
        self.ekle_pencere = tk.Toplevel(self.ana_pencere)
        self.ekle_pencere.title("Tarif Ekle")

        self.etiket = tk.Label(self.ekle_pencere, text="Tarifinizi girin:")
        self.etiket.pack()

        self.tarif_metni = tk.Text(self.ekle_pencere, height=10, width=50)
        self.tarif_metni.pack()

        self.kaydet_btn = tk.Button(self.ekle_pencere, text="Kaydet", command=self.tarifi_kaydet)
        self.kaydet_btn.pack()

    def tarifi_kaydet(self):
        tarif = self.tarif_metni.get("1.0", tk.END)
        with open("tarifler.txt", "a") as dosya:
            dosya.write(tarif + "\n")
        self.ekle_pencere.destroy()

    def tarif_ara(self):
        self.ara_pencere = tk.Toplevel(self.ana_pencere)
        self.ara_pencere.title("Tarif Arama")

        self.etiket = tk.Label(self.ara_pencere, text="Mevcut Tarifler:")
        self.etiket.pack()

        with open("tarifler.txt", "r") as dosya:
            tarifler = dosya.readlines()
            for tarif in tarifler:
                self.etiket = tk.Label(self.ara_pencere, text=tarif.strip())
                self.etiket.pack()

    def tarif_degerlendir(self):
        self.degerlendir_pencere = tk.Toplevel(self.ana_pencere)
        self.degerlendir_pencere.title("Tarif Değerlendirme")

        self.etiket = tk.Label(self.degerlendir_pencere, text="Bu tarifi beğendiniz mi?")
        self.etiket.pack()

        self.beğen_btn = tk.Button(self.degerlendir_pencere, text="Beğendim", command=lambda: self.degerlendir_tarif(True))
        self.beğen_btn.pack()

        self.begenme_btn = tk.Button(self.degerlendir_pencere, text="Beğenmedim", command=lambda: self.degerlendir_tarif(False))
        self.begenme_btn.pack()

    def degerlendir_tarif(self, begenildi_mi):
        degerlendirme = "Beğendim" if begenildi_mi else "Beğenmedim"
        with open("degerlendirmeler.txt", "a") as dosya:
            dosya.write(degerlendirme + "\n")
        self.degerlendir_pencere.destroy()

ana_pencere = tk.Tk()
uygulama = TarifUygulamasi(ana_pencere)
ana_pencere.mainloop()
```
