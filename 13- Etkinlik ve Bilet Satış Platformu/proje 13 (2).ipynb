```python
import tkinter as tk

class EventTicketingApp:
    def __init__(self, master):
        self.master = master
        self.master.title("Etkinlik ve Bilet Satış Platformu")

        # Kullanıcı adı giriş kutuları
        self.first_name_label = tk.Label(self.master, text="İsim:")
        self.first_name_label.grid(row=0, column=1, padx=10, pady=5)
        self.first_name_entry = tk.Entry(self.master)
        self.first_name_entry.grid(row=0, column=2, padx=10, pady=5)

        self.last_name_label = tk.Label(self.master, text="Soyisim:")
        self.last_name_label.grid(row=1, column=1, padx=10, pady=5)
        self.last_name_entry = tk.Entry(self.master)
        self.last_name_entry.grid(row=1, column=2, padx=10, pady=5)

        # Etkinliklerin listeleneceği bir liste kutusu
        self.event_listbox = tk.Listbox(self.master, width=50)
        self.event_listbox.grid(row=2, column=0, padx=10, pady=10)

        # Bilet satın alma düğmesi
        self.buy_ticket_button = tk.Button(self.master, text="Bilet Satın Al", command=self.buy_ticket)
        self.buy_ticket_button.grid(row=3, column=0, padx=10, pady=5)

        # Biletlerin listeleneceği bir liste kutusu
        self.ticket_listbox = tk.Listbox(self.master, width=50)
        self.ticket_listbox.grid(row=4, column=0, padx=10, pady=10)

        # Bilet iptal düğmesi
        self.cancel_ticket_button = tk.Button(self.master, text="Bileti İptal Et", command=self.cancel_ticket)
        self.cancel_ticket_button.grid(row=5, column=0, padx=10, pady=5)

        # Etkinlikleri yükle (örnek veri)
        self.load_events()

    def load_events(self):
        # Veritabanından etkinlikleri alıp listbox'a ekleyen bir işlev olabilir
        events = ["Konser - 10.05.2024", "Tiyatro Gösterisi - 15.05.2024", "Spor Müsabakası - 20.05.2024"]
        for event in events:
            self.event_listbox.insert(tk.END, event)

    def buy_ticket(self):
        selected_event = self.event_listbox.get(tk.ACTIVE)
        first_name = self.first_name_entry.get()
        last_name = self.last_name_entry.get()
        # Kullanıcı adını ve soyadını alıp bilet satın alma işlemleri yapılabilir
        # Bilet satın alma işlemleri tamamlandıktan sonra bilet_listbox'a ekleyebilirsiniz
        ticket_info = f"{first_name} {last_name}: {selected_event}"
        self.ticket_listbox.insert(tk.END, ticket_info)

    def cancel_ticket(self):
        selected_ticket = self.ticket_listbox.get(tk.ACTIVE)
        # Seçilen bilet için iptal işlemleri yapılabilir
        # İptal işlemi gerçekleştikten sonra bilet_listbox'tan silinebilir
        self.ticket_listbox.delete(tk.ACTIVE)

def main():
    root = tk.Tk()
    app = EventTicketingApp(root)
    root.mainloop()

if __name__ == "__main__":
    main()
```
