```python
import tkinter as tk
from tkinter import messagebox

class Egitmen:
    def __init__(self, isim, uzmanlik_alani):
        self.isim = isim
        self.uzmanlik_alani = uzmanlik_alani

class Ogrenci:
    def __init__(self, isim, email):
        self.isim = isim
        self.email = email

class Kurs:
    def __init__(self, kurs_adi, egitmen, icerik):
        self.kurs_adi = kurs_adi
        self.egitmen = egitmen
        self.icerik = icerik
        self.ogrenciler = []

    def kurs_olustur(self):
        # Kurs oluşturma işlemleri
        pass

    def kaydol(self, ogrenci):
        self.ogrenciler.append(ogrenci)

    def icerik_yukle(self):
        # Kurs içeriği yükleme işlemleri
        pass

class KursArayuzu:
    def __init__(self, kurslar):
        self.kurslar = kurslar
        self.ana_pencere = tk.Tk()
        self.ana_pencere.title("Kurs Arayüzü")

        self.kurs_listesi = tk.Listbox(self.ana_pencere, width=50)
        self.kurs_listesi.pack(pady=10)

        for kurs in self.kurslar:
            self.kurs_listesi.insert(tk.END, kurs.kurs_adi)

        self.kaydol_button = tk.Button(self.ana_pencere, text="Kaydol", command=self.kaydol)
        self.kaydol_button.pack(pady=5)

        self.icerik_goster_button = tk.Button(self.ana_pencere, text="İçerik Göster", command=self.icerik_goster)
        self.icerik_goster_button.pack(pady=5)

        self.isim_label = tk.Label(self.ana_pencere, text="İsim:")
        self.isim_label.pack()
        self.isim_entry = tk.Entry(self.ana_pencere)
        self.isim_entry.pack()

        self.email_label = tk.Label(self.ana_pencere, text="E-posta:")
        self.email_label.pack()
        self.email_entry = tk.Entry(self.ana_pencere)
        self.email_entry.pack()

    def kaydol(self):
        selected_kurs_index = self.kurs_listesi.curselection()
        if selected_kurs_index:
            selected_kurs = self.kurslar[selected_kurs_index[0]]
            ogrenci_isim = self.isim_entry.get()
            ogrenci_email = self.email_entry.get()
            if ogrenci_isim and ogrenci_email:
                ogrenci = Ogrenci(ogrenci_isim, ogrenci_email)
                selected_kurs.kaydol(ogrenci)
                messagebox.showinfo("Başarılı", f"{ogrenci_isim}, {selected_kurs.kurs_adi} kursuna başarıyla kaydoldunuz!")
            else:
                messagebox.showerror("Hata", "Lütfen isim ve e-posta adresinizi giriniz.")
        else:
            messagebox.showerror("Hata", "Lütfen bir kurs seçiniz.")

    def icerik_goster(self):
        selected_kurs_index = self.kurs_listesi.curselection()
        if selected_kurs_index:
            selected_kurs = self.kurslar[selected_kurs_index[0]]
            icerik = selected_kurs.icerik
            messagebox.showinfo("Kurs İçeriği", icerik)
        else:
            messagebox.showerror("Hata", "Lütfen bir kurs seçiniz.")

# Örnek veri oluşturma
egitmen1 = Egitmen("Ahmet Bey", "Python Programlama")
egitmen2 = Egitmen("Ayşe Hanım", "Veri Bilimi")

kurs1 = Kurs("Python Programlama", egitmen1, "Temel Python dersleri")
kurs2 = Kurs("Veri Bilimi", egitmen2, "Veri analizi ve makine öğrenmesi konularını içerir")

kurslar = [kurs1, kurs2]

# Kullanım
if __name__ == "__main__":
    arayuz = KursArayuzu(kurslar)
    arayuz.ana_pencere.mainloop()
```
