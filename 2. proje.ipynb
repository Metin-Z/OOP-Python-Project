```python
import tkinter as tk
from tkinter import messagebox

class Hasta:
    def __init__(self, isim, tc):
        self.isim = isim
        if len(tc) == 11:
            self.tc = tc
        else:
            raise ValueError("TC kimlik numarası 11 haneli olmalıdır.")

class Doktor:
    def __init__(self, isim, uzmanlik_alani):
        self.isim = isim
        self.uzmanlik_alani = uzmanlik_alani
        self.musaitlik_durumu = True

class Randevu:
    def __init__(self, randevu_tarihi, doktor, hasta):
        self.randevu_tarihi = randevu_tarihi
        self.doktor = doktor
        self.hasta = hasta

    def randevu_al(self):
        if self.doktor.musaitlik_durumu:
            self.doktor.musaitlik_durumu = False
            messagebox.showinfo("Başarılı", "Randevu başarıyla alındı.")
        else:
            messagebox.showwarning("Uyarı", "Doktorunuz müsait değil.")

    def randevu_iptal(self):
        self.doktor.musaitlik_durumu = True
        messagebox.showinfo("Başarılı", "Randevu başarıyla iptal edildi.")

class RandevuUygulamasi:
    def __init__(self, master):
        self.master = master
        self.master.title("Randevu Sistemi")

        self.label_hasta_isim = tk.Label(master, text="Hasta İsim:")
        self.label_hasta_isim.grid(row=0, column=0, padx=5, pady=5)
        self.entry_hasta_isim = tk.Entry(master)
        self.entry_hasta_isim.grid(row=0, column=1, padx=5, pady=5)

        self.label_hasta_tc = tk.Label(master, text="Hasta TC:")
        self.label_hasta_tc.grid(row=1, column=0, padx=5, pady=5)
        self.entry_hasta_tc = tk.Entry(master)
        self.entry_hasta_tc.grid(row=1, column=1, padx=5, pady=5)

        self.label_doktor_isim = tk.Label(master, text="Doktor İsim:")
        self.label_doktor_isim.grid(row=2, column=0, padx=5, pady=5)
        self.entry_doktor_isim = tk.Entry(master)
        self.entry_doktor_isim.grid(row=2, column=1, padx=5, pady=5)

        self.label_randevu_tarih = tk.Label(master, text="Randevu Tarihi:")
        self.label_randevu_tarih.grid(row=3, column=0, padx=5, pady=5)
        self.entry_randevu_tarih = tk.Entry(master)
        self.entry_randevu_tarih.grid(row=3, column=1, padx=5, pady=5)

        self.label_musaitlik_durumu = tk.Label(master, text="Doktorun Müsaitlik Durumu: Müsait")
        self.label_musaitlik_durumu.grid(row=4, column=0, columnspan=2, padx=5, pady=5)

        self.btn_randevu_al = tk.Button(master, text="Randevu Al", command=self.randevu_al)
        self.btn_randevu_al.grid(row=5, column=0, padx=5, pady=5)

        self.btn_randevu_iptal = tk.Button(master, text="Randevu İptal", command=self.randevu_iptal)
        self.btn_randevu_iptal.grid(row=5, column=1, padx=5, pady=5)

    def randevu_al(self):
        hasta_isim = self.entry_hasta_isim.get()
        hasta_tc = self.entry_hasta_tc.get()
        doktor_isim = self.entry_doktor_isim.get()
        randevu_tarih = self.entry_randevu_tarih.get()

        try:
            hasta = Hasta(hasta_isim, hasta_tc)
            doktor = Doktor(doktor_isim, "Uzmanlık Alanı")
            randevu = Randevu(randevu_tarih, doktor, hasta)

            randevu.randevu_al()
            self.guncelle_musaitlik_durumu(doktor)
        except ValueError as e:
            messagebox.showerror("Hata", str(e))

    def randevu_iptal(self):
        messagebox.showinfo("Uyarı", "Randevu iptal işlemi henüz kodlanmadı!")

    def guncelle_musaitlik_durumu(self, doktor):
        if doktor.musaitlik_durumu:
            self.label_musaitlik_durumu.config(text="Doktorun Müsaitlik Durumu: Müsait", fg="green")
        else:
            self.label_musaitlik_durumu.config(text="Doktorun Müsaitlik Durumu: Meşgul", fg="red")

def main():
    root = tk.Tk()
    app = RandevuUygulamasi(root)
    root.mainloop()

if __name__ == "__main__":
    main()
```
